```python
# Load the dataset
df=pd.read_csv("../../data/raw/creditcard.csv")
```

```python
df.head()
```

## Dataset Description
- Dataset contains only numerical input variables which are the result of a PCA transformation. 
- Unfortunately, due to confidentiality issues, the dataset does not contain the original features and more background information about the data. 
- Features V1, V2, … V28 are the principal components obtained with PCA, the only features which have not been transformed with PCA are 'Time' and 'Amount'. 
- Feature 'Time' contains the seconds elapsed between each transaction and the first transaction in the dataset. 
- The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost-sensitive learning. 
- Feature 'Class' is the response variable and it takes value 1 in case of fraud and 0 otherwise.


- `Id` - A unique Id for each row.
- `Time` - Number of seconds elapsed between this transaction and the first transaction in the dataset
- `V1-V28` - Features after dimensionality reduction to protect user identities and sensitive features
- `Amount` - Transaction amount
- `Class` - Target Class (1 for fraudulent transactions, 0 genuine)

```python
numeric_columns = (list(df.loc[:, 'Time':'Amount']))

fig = plt.figure(figsize=(20, 50))
rows, cols = 10, 3
for idx, num in enumerate(numeric_columns[:30]):
    ax = fig.add_subplot(rows, cols, idx+1)
    ax.grid(alpha = 0.7, axis ="both")
    sns.kdeplot(x = num, fill = True,color ="#3386FF",linewidth=0.6, data = df, label = "Dataset") 
    ax.set_xlabel(num)
    ax.legend()
fig.tight_layout()
```

## Key Observations
- The dataset consists of PCA-transformed features (V1–V28), which are mostly centered around zero and approximately normally distributed.
- Several features exhibit heavy tails and outliers, which may contain important fraud-related signals and should not be removed.
- The Amount feature is highly right-skewed and requires scaling or log transformation.
- The Time feature shows non-uniform distribution but is not strongly informative in isolation.
- Minimal preprocessing is required due to prior transformation of features.

```python
df=df.drop('Time',axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
df['Amount'] = np.log1p(df['Amount'])

scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df[['Amount']])
```

```python
sns.kdeplot(x = 'Amount', fill = True,color ="#3386FF",linewidth=0.6, data = df, label = "Dataset") 
```

```python
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (Fraud vs Non-Fraud)")
plt.show()
```

```python
df['Class'].value_counts().plot.pie(autopct='%1.2f%%', labels=['Legit', 'Fraud'], colors=['skyblue','red'])
plt.title("Fraud Class Imbalance")
plt.ylabel("")
plt.show()

```

## Class Imbalance in the Dataset

- The dataset exhibits a severe class imbalance, with approximately 99.83% legitimate transactions and only 0.17% fraudulent transactions.

- This imbalance presents a significant challenge for model training, as standard classifiers may become biased toward predicting the majority class, resulting in poor detection of fraudulent cases.

- To address this, the modeling phase will incorporate:

  - Class-weighted learning
  - Resampling techniques (e.g., SMOTE)
  - Threshold optimization to balance precision and recall

## Conclusion

- Dataset is PCA-transformed → minimal preprocessing required
- Amount is skewed → log + scaling needed
- Severe class imbalance → main challenge
- Focus shifts to modeling techniques rather than feature engineering
